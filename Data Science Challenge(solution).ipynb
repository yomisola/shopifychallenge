a. The Average order value given was gotten by getting the average  of the order_amount as seen above, this would have been appropriate if the order_amount and total_items were consistent.

Looking deeper into the dataset, I realized that there need to be some sort of grouping by quantity unit which is total items in this case.

I grouped the unit purchase by individual buyers(=<20 items) and by wholesale buyers(>20items) before getting their averages with the describe function.


b. The metric I would report is the average Order amount(AOV) for the individual buyers and the wholesale buyers.

c. The values are 
      
      
      i.  $745.09
      
      ii. $704000 respectively.

```python
b2b = []
for i in df.total_items:
    if i > 20:
        b2b.append(True)
    else:
        b2b.append(False)

df['is_b2b'] = b2b
print(df.shape)
df.sample(100)
```

```python
#Dataframe for the individual buyers
df2 = df[~df['is_b2b']]
print(df2.shape)
df2.sample(10)
```

```python
#Dataframe for the wholesale buyers
df3 = df[df['is_b2b']]
print(df3.shape)
df3.sample(10)
```
